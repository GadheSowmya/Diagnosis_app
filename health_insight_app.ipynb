```python
@app.route("/predict", methods=["POST"])
def predict_api():
    data = request.json
    text = data.get("description", "").lower()

    hormonal_score = 0
    diabetes_score = 0
    thyroid_score = 0

    hormonal_keywords = [
        "irregular", "period", "acne", "hair growth",
        "weight gain", "pcod", "pcos", "hormone"
    ]

    diabetes_keywords = [
        "thirst", "urination", "sugar",
        "diabetes", "blood sugar", "fatigue", "tired"
    ]

    thyroid_keywords = [
        "thyroid", "hair fall", "weight change",
        "cold", "heat", "slow metabolism", "neck swelling"
    ]

    for w in hormonal_keywords:
        if w in text:
            hormonal_score += 1

    for w in diabetes_keywords:
        if w in text:
            diabetes_score += 1

    for w in thyroid_keywords:
        if w in text:
            thyroid_score += 1

    if hormonal_score == 0 and diabetes_score == 0 and thyroid_score == 0:
        return jsonify(PREVENTIVE)

    if thyroid_score >= hormonal_score and thyroid_score >= diabetes_score:
        return jsonify(RESPONSES[3])

    if diabetes_score > hormonal_score:
        return jsonify(RESPONSES[2])

    return jsonify(RESPONSES[0])
```
